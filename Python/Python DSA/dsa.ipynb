```python
import pandas as pd

data = {
    "date": [
        "2024-01-01", "2024-01-02", "2024-01-03",
        "2024-01-04", "2024-01-05", "2024-01-06","2024-01-06"
    ],
    "amount": [50, 120, 75, 200, 95, 300, None],
    "country": ["UK", "USA", "UK", "UK", "Canada", "USA", None]
}

df = pd.DataFrame(data)
print(df)
```
